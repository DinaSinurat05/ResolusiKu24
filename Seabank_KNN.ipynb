```python
#clenning
#remove url
df0012_data['content'] = df0012_data['content'].str.replace('http\s',' ',case=False)

#merubah teks menjdi huruf kecil
df0012_data['content'] = df0012_data['content'].str.lower()

# Remove mention (@username)
df0012_data['content'] = df0012_data['content'].str.replace('@\s+', ' ',case=False)

# Remove hashtag (#hashtag)
df0012_data['content'] = df0012_data['content'].str.replace('#\s+', ' ', case=False)

# Remove new line character
df0012_data['content'] = df0012_data['content'].str.replace('\w+', ' ', case=False)

# Remove punctuation
df0012_data['content'] = df0012_data['content'].str.replace('[^\w\s]', ' ', case=False)

# Remove numbers
df0012_data['content'] = df0012_data['content'].str.replace('w*\d+*', ' ', case=False)

# Remove spasi berlebihan
df0012_data['content'] = df0012_data['content'].str.replace('\s(2)', ' ', case=False)


```

```python
#tokenize
#resting
from nltk.tokenize import word_tokenize

x = df0012_data.iloc[0]
print(word_tokenize(x['content']))
```

```python
def identify_tokens(row):
  text = row['content']
  tokens =nltk.word_tokenize(text)
  token_words = [w for w in tokens if w.isalpha()]
  return token_words

df0012_data['content'] = df0012_data.apply(identify_tokens, axis=1)
df0012_data.content
```

```python
#stemming pembentukan kata dasar
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
stemming = PorterStemmer()
```

```python
def stem_list(row):
  text = row['content']
  stem = [stemming.stem(word) for word in text]
  return(stem)

df0012_data['content'] = df0012_data.apply(stem_list, axis= 1)
df0012_data.content
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd

# Mengambil konten teks sebagai Series
content = df0012_data['content'].astype(str)  # Konversi ke string jika belum dalam format string

# Menginisialisasi CountVectorizer
tfidfvectorizer = CountVectorizer(analyzer='word')

# Menghitung vektor
tfid_wm = tfidfvectorizer.fit_transform(content)
tfid_tokens = tfidfvectorizer.get_feature_names_out()

# Mengonversi hasil menjadi DataFrame
df_tfidfvec = pd.DataFrame(tfid_wm.toarray(), columns=tfid_tokens)

print("\nTF-IDF Vectorizer")
print(df_tfidfvec)
```
